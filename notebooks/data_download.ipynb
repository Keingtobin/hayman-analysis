```python
import os
import requests
import pathlib
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
# Define and create all data directories.
ndvi_base_url = "https://www.ncei.noaa.gov/data/land-normalized-difference-vegetation-index/access/"
project_directory = pathlib.Path.cwd().parent
data_directory = project_directory / "data"
os.makedirs(data_directory, exist_ok=True)
NDVI_directory = data_directory / "NDVI"
os.makedirs(NDVI_directory, exist_ok=True)
```

```python
# Get all years we want.
resp = requests.get(ndvi_base_url)
soup = BeautifulSoup(resp.text, "html.parser")
years = []
# Get only years we want (2002 to now).
for link in soup.find_all("a"):
    href = link.get("href")
    if href.isdigit():
        href = int(href)
        if href >= 2001:
            years.append(href)
```

```python
# Download all july data for each year.
for year in years:
    year_dir = NDVI_directory / str(year)
    os.makedirs(year_dir, exist_ok=True)
    year_url = f'{ndvi_base_url}{year}/'
    resp = requests.get(year_url)
    soup = BeautifulSoup(resp.text, "html.parser")
    for link in soup.find_all("a"):
        file_name = link.get("href")
        if file_name and f'{year}07' in file_name:
            file_url = f'{year_url}{file_name}'
            local_filepath = os.path.join(year_dir, file_name)
            if os.path.exists(local_filepath):
                print(f'Skipping existing file: {local_filepath}')
                continue
            print(f'Downloading {file_url}')
            with requests.get(file_url, stream=True) as r:
                r.raise_for_status()
                with open(local_filepath, "wb") as f:
                    for chunk in r.iter_content(chunk_size=8192):
                        f.write(chunk)
```
